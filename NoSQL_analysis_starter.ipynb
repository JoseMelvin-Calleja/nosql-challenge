## Notebook Set Up

```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
uk_food = mongo['uk_food']
```

```python
# review the collections in our database
uk_food.list_collection_names()
```

```python
# assign the collection to a variable
establishments = uk_food['establishments']
```

## Part 3: Exploratory Analysis
Unless otherwise stated, for each question:

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": {'$eq': 20}}
results = establishments.find(query)

# Use count_documents to display the number of documents in the result
hygiene_docs = establishments.count_documents(query)
print(f"There are {hygiene_docs} establishments with a Hygiene score of 20.")

# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
hygiene_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f"There are {len(hygiene_df)} rows in the DataFrame")

# Display the first 10 rows of the DataFrame
hygiene_df.head()
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# changing RatingValue from String to Int
establishments.update_many({}, [{'$set':
                                    {'RatingValue': {'$toInt': '$RatingValue'}}
                                }])
```

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {'RatingValue': {'$gte': 4}}
results = establishments.find(query)

# Use count_documents to display the number of documents in the result
rating_value_docs = establishments.count_documents(query)
print(f"There are {rating_value_docs} establishments with a RatingValue greater than or equal to 4.")

# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
rating_value_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f"Ther are {len(rating_value_df)} rows in the DataFrame.")

# Display the first 10 rows of the DataFrame
rating_value_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = establishments.find_one({'BusinessName': 'Penang Flavours'}, {'geocode.latitude': 1})['geocode']['latitude']
longitude = establishments.find_one({'BusinessName': 'Penang Flavours'}, {'geocode.longitude': 1})['geocode']['longitude']

query = {
    'RatingValue': {'$eq': 5}, 
    'geocode.latitude': {'$lte': latitude + degree_search, '$gte': latitude - degree_search},
    'geocode.longitude': {'$lte': longitude + degree_search, '$gte': longitude - degree_search}}
sort =  [('scores.Hygiene', 1)]

# Print the results
results = list(establishments.find(query).limit(5).sort(sort))

pprint(results)
```

```python
# Convert result to Pandas DataFrame
near_penang_df = pd.DataFrame(results)
near_penang_df
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match_query = {'$match': {'scores.Hygiene': {'$eq': 0}}}

group_query = {'$group': {'_id': '$LocalAuthorityName', 'count': {'$sum': 1}}}

sort_query = {'$sort': {'count': -1}}

query = [match_query, group_query, sort_query]

results = list(establishments.aggregate(query))

# Print the number of documents in the result
print(f"There are {len(results)} documents")
# Print the first 10 results
pprint(results[0:9])
```

```python
# Convert the result to a Pandas DataFrame
local_authority_hygiene_df = pd.DataFrame(results)
# Display the number of rows in the DataFrame
print(len(local_authority_hygiene_df))
# Display the first 10 rows of the DataFrame
local_authority_hygiene_df.head(10)
```
